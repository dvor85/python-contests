## 1. –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç nnn –ø–µ—Ä–≤—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ 1‚Äâ2‚Äâ2‚Äâ3‚Äâ3‚Äâ3‚Äâ4‚Äâ4‚Äâ4‚Äâ4‚Äâ5‚Äâ5‚Äâ5‚Äâ5‚Äâ5‚Ä¶(—á–∏—Å–ª–æ –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è —Å—Ç–æ–ª—å–∫–æ —Ä–∞–∑, —á–µ–º—É –æ–Ω–æ —Ä–∞–≤–Ω–æ).

```python
def solution(n):
    l = 0
    for i in range(1, n + 1):
        l += i
        r = i if n - l > 0 else i + n - l
        if r > 0:
            print(' '.join([str(i)] * r), end=' ')
        else:
            return


n=int(input())        
solution(n)
```

## 2. BeeGeek
–†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é beegeek(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–∞ aaa –∏ b, –≥–¥–µ a‚â§b, –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É, —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—É—é –∏–∑ —á–∏—Å–µ–ª –æ—Ç a –¥–æ b –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –∏–ª–∏ —Å–ª–æ–≤ Bee, Geek –∏ BeeGeek –ø–æ —Å–ª–µ–¥—É—é—â–µ–º—É –ø—Ä–∞–≤–∏–ª—É:

    –µ—Å–ª–∏ —á–∏—Å–ª–æ –¥–µ–ª–∏—Ç—Å—è –±–µ–∑ –æ—Å—Ç–∞—Ç–∫–∞ –Ω–∞ 3, —Ç–æ –≤–º–µ—Å—Ç–æ –Ω–µ–≥–æ –≤ —Å—Ç—Ä–æ–∫—É –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è —Å–ª–æ–≤–æ Bee;
    –µ—Å–ª–∏ —á–∏—Å–ª–æ –¥–µ–ª–∏—Ç—Å—è –±–µ–∑ –æ—Å—Ç–∞—Ç–∫–∞ –Ω–∞ 7, —Ç–æ –≤–º–µ—Å—Ç–æ –Ω–µ–≥–æ –≤ —Å—Ç—Ä–æ–∫—É –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è —Å–ª–æ–≤–æ Geek;
    –µ—Å–ª–∏ —á–∏—Å–ª–æ –¥–µ–ª–∏—Ç—Å—è –±–µ–∑ –æ—Å—Ç–∞—Ç–∫–∞ –∏ –Ω–∞ 3, –∏ –Ω–∞ 7, —Ç–æ –≤–º–µ—Å—Ç–æ –Ω–µ–≥–æ –≤ —Å—Ç—Ä–æ–∫—É –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è —Å–ª–æ–≤–æ BeeGeek;
    –≤ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å–ª—É—á–∞—è—Ö –≤ —Å—Ç—Ä–æ–∫—É –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è —Å–∞–º–æ —á–∏—Å–ª–æ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –ß–∏—Å–ª–∞ –∏ —Å–ª–æ–≤–∞ –≤ —Ñ–æ—Ä–º–∏—Ä—É–µ–º–æ–π —Å—Ç—Ä–æ–∫–µ –¥–æ–ª–∂–Ω—ã —Ä–∞–∑–¥–µ–ª—è—Ç—å—Å—è —Å–∏–º–≤–æ–ª–æ–≤ –ø—Ä–æ–±–µ–ª–∞.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –í —Ç–µ—Å—Ç–∏—Ä—É—é—â—É—é —Å–∏—Å—Ç–µ–º—É —Å–¥–∞–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, —Å–æ–¥–µ—Ä–∂–∞—â—É—é —Ç–æ–ª—å–∫–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é —Ñ—É–Ω–∫—Ü–∏—é beegeek(), –Ω–æ –Ω–µ –∫–æ–¥, –≤—ã–∑—ã–≤–∞—é—â–∏–π –µ–µ.

```python
def beegeek(a, b):
    res = []
    for i in range(a,b+1):
        if (i % 3 == 0) and (i % 7 == 0):
            res.append('BeeGeek')
        elif i % 3 == 0:
            res.append('Bee')
        elif i % 7 == 0:
            res.append('Geek')
        else:
            res.append(str(i))
            
    return ' '.join(res)

beegeek(10,50)
```

## 3. –í–∑–ª–æ–º –≤—Ä–∞—Ç
–ë–µ–∑—É–º–Ω—ã–π —É—á–µ–Ω—ã–π –∏–∑–æ–±—Ä–µ–ª –º–∞—à–∏–Ω—É –≤—Ä–µ–º–µ–Ω–∏, –ø–æ–¥–∫–ª—é—á–∏–≤ –º–∏–∫—Ä–æ–≤–æ–ª–Ω–æ–≤–∫—É –∫ —Ç–µ–ª–µ—Ñ–æ–Ω—É, –∏ —Å –ø–æ–º–æ—â—å—é —ç—Ç–æ–≥–æ —Ç–µ–ª–µ—Ñ–æ–Ω–∞ –æ—Ç–ø—Ä–∞–≤–ª—è–µ—Ç –ø–∏—Å—å–º–∞ –≤ –ø—Ä–æ—à–ª–æ–µ. –û–¥–Ω–∞–∫–æ –æ–Ω–∞ –ø–æ–ª—É—á–∏–ª–∞—Å—å –Ω–∞—Å—Ç–æ–ª—å–∫–æ —É–∂–∞—Å–Ω–æ–π, —á—Ç–æ –º–∞–ª–æ —Ç–æ–≥–æ —á—Ç–æ –≤ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–º —Å–æ–æ–±—â–µ–Ω–∏–∏ –º–æ–≥—É—Ç –ø–µ—Ä–µ–º–µ—à–∏–≤–∞—Ç—å—Å—è —Å–∏–º–≤–æ–ª—ã, —Ç–∞–∫ –µ—â–µ –∏ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –æ–¥–∏–Ω –ª–∏—à–Ω–∏–π. –ù–æ, –µ—Å–ª–∏ –≤—ã –Ω–∞–π–¥–µ—Ç–µ —ç—Ç–æ—Ç —Å–∏–º–≤–æ–ª, –±—ã—Ç—å –º–æ–∂–µ—Ç, –Ω–µ–∏—Å–ø—Ä–∞–≤–Ω–æ—Å—Ç—å –±—É–¥–µ—Ç –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∞? 

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –ª–∏—à–Ω–∏–π —Å–∏–º–≤–æ–ª –≤ –∏–∑–º–µ–Ω–µ–Ω–Ω–æ–º —Å–æ–æ–±—â–µ–Ω–∏–∏.

```python
def solution(a, b):
    sa, sb = sorted(a), sorted(b)
    for i, s in enumerate(sb):
        if i >= len(sa) or s != sa[i]:
            return s


solution('üòÅüòÇüòÉüòÑüòÖüòÜüòáüòàüòâüòäüòãüòåüòçüòé', 'üòÅüòâüòÇüòÉüòÑüòáüòÖüòÜüòàüòäüòãüòåüòç‚ú®üòé')
```

## 4. NRZI –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
NRZI –∫–æ–¥ (Non Return to Zero Invertive) ‚Äî –æ–¥–∏–Ω –∏–∑ —Å–ø–æ—Å–æ–±–æ–≤ –ª–∏–Ω–µ–π–Ω–æ–≥–æ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è. –°—É—Ç—å —ç—Ç–æ–≥–æ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –≤ —Ç–æ–º, —á—Ç–æ –∏–º–µ—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ, –∏–º–µ—é—â–µ–µ –≤—Å–µ–≥–æ –¥–≤–∞ —Å–æ—Å—Ç–æ—è–Ω–∏—è, –º—ã —Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É –µ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏–π –Ω–∞ –∫–∞–∂–¥–æ–º —Ç–∞–∫—Ç–µ, –∏ –µ—Å–ª–∏ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –∏–∑–º–µ–Ω–∏–ª–æ—Å—å, —Ç–æ —ç—Ç–æ —Ä–∞—Å—Ü–µ–Ω–∏–≤–∞–µ—Ç—Å—è –∫–∞–∫ –¥–≤–æ–∏—á–Ω–∞—è –µ–¥–∏–Ω–∏—Ü–∞, –µ—Å–ª–∏ –∂–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –æ—Å—Ç–∞–ª–æ—Å—å –Ω–µ–∏–∑–º–µ–Ω–Ω—ã–º, —Ç–æ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –¥–≤–æ–∏—á–Ω—ã–π –Ω–æ–ª—å.
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—Ä–µ–≤–æ–¥–∏—Ç NRZI –∫–æ–¥ –≤ –¥–≤–æ–∏—á–Ω—ã–π.

```python
import re


def solution(nrzi):
    ones = re.compile(r'\|[¬Ø_]')
    zeros = re.compile(r'[¬Ø_]')
    return zeros.sub('0', ones.sub('1', nrzi))


solution('|¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø¬Ø|_|¬Ø|____|¬Ø|_')
```

## 5. –ü–æ—á—Ç–∏ –ø–∞–ª–∏–Ω–¥—Ä–æ–º
–ù–∞–∑–æ–≤–µ–º —Å—Ç—Ä–æ–∫—É —Ç–µ–∫—Å—Ç–∞ ¬´–ø–æ—á—Ç–∏ –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º¬ª, –µ—Å–ª–∏ –Ω–∞–π–¥–µ—Ç—Å—è —Ç–∞–∫–æ–π –±—É–∫–≤–µ–Ω–Ω—ã–π —Å–∏–º–≤–æ–ª, –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ —Å—Ç—Ä–æ–∫–∞ —Å—Ç–∞–Ω–µ—Ç –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º. –ü—Ä–∏ —ç—Ç–æ–º –≤—Å–µ —Å–∏–º–≤–æ–ª—ã, –∫—Ä–æ–º–µ –±—É–∫–≤, –¥–æ–ª–∂–Ω—ã –∏–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞—Ç—å—Å—è.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Å—Ç—Ä–æ–∫–∞ ¬´–ø–æ—á—Ç–∏ –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º¬ª.

```python
import re


def check(a):
    return a == a[::-1]


def solution(pali):
    w = re.sub(r'[^a-z]', '', pali)
    for i in range(len(w) - 1):
        if (i >= len(w) - i - 1):
            return True
        if w[i] != w[-i - 1]:
            return check(w[i + 1:-i]) or check(w[i:-i - 1])

    return True


solution('1kilg%rli8k')
```

## 6. –û–≥–æ—Ä–æ–¥–Ω—ã–π –ø–æ–∫–µ—Ä
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å –Ω–∞–∏–ª—É—á—à—É—é –ø–æ–∫–µ—Ä–Ω—É—é –∫–æ–º–±–∏–Ω–∞—Ü–∏—é –≤ –¥–∞–Ω–Ω–æ–π —Ä—É–∫–µ –∏–∑ 5 –∫–∞—Ä—Ç. –í–∞–ª–µ—Ç—ã, –¥–∞–º—ã, –∫–æ—Ä–æ–ª–∏ –∏ —Ç—É–∑—ã –±—É–¥—É—Ç –¥–∞–Ω—ã –∫–∞–∫ —á–∏—Å–ª–∞ 11, 12, 13 –∏ 1 —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

–í –ª—é–±–æ–π –∫–∞—Ä—Ç–æ—á–Ω–æ–π –∏–≥—Ä–µ –µ—Å—Ç—å –º–µ—Å—Ç–æ —à—É–ª–µ—Ä—Å—Ç–≤—É, –ø–æ—ç—Ç–æ–º—É —Å–ª–µ–¥—É–µ—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Ä—É–∫—É –Ω–∞ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –µ–µ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è. –í —Ä—É–∫–µ –∏–∑ –ø—è—Ç–∏ –∫–∞—Ä—Ç –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –±–æ–ª–µ–µ 4 –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö –∫–∞—Ä—Ç, –µ—Å–ª–∏ —Ç–∞–∫–æ–µ –ø—Ä–æ–∏–∑–æ—à–ª–æ ‚Äì —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ —Å–ª–æ–≤–æ –®—É–ª–µ—Ä.

–ö–æ–º–±–∏–Ω–∞—Ü–∏–∏ –ø–æ —É–±—ã–≤–∞–Ω–∏—é —Å—Ç–∞—Ä—à–∏–Ω—Å—Ç–≤–∞:

    4 –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∫–∞—Ä—Ç—ã ‚Äì –≤—ã–≤–µ—Å—Ç–∏ –ö–∞—Ä–µ;
    3 –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∫–∞—Ä—Ç—ã –∏ 2 –¥—Ä—É–≥–∏–µ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∫–∞—Ä—Ç—ã ‚Äì –≤—ã–≤–µ—Å—Ç–∏ –§—É–ª–ª –•–∞—É—Å;
    5 –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –∏–¥—É—â–∏—Ö –∫–∞—Ä—Ç ‚Äì –°—Ç—Ä–∏—Ç;
    3 –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∫–∞—Ä—Ç—ã ‚Äì –°–µ—Ç;
    2 –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∫–∞—Ä—Ç—ã –∏ 2 –¥—Ä—É–≥–∏–µ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∫–∞—Ä—Ç—ã ‚Äì –î–≤–µ –ø–∞—Ä—ã;
    2 –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –∫–∞—Ä—Ç—ã ‚Äì –ü–∞—Ä–∞;
    –Ω–∏—á–µ–≥–æ –∏–∑ –≤—ã—à–µ–ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ ‚Äì –°—Ç–∞—Ä—à–∞—è –∫–∞—Ä—Ç–∞.


```python
def solution(scards):
    cards = sorted(map(int, scards.split()))
    c_cards = [cards.count(c) for c in set(cards)]

    if len(c_cards) == 1:
        return '–®—É–ª–µ—Ä'
    if cards == list(range(cards[0], cards[0] + 5)):
        return '–°—Ç—Ä–∏—Ç'
    if 4 in c_cards:
        return '–ö–∞—Ä–µ'
    if 3 in c_cards:
        if 2 in c_cards:
            return '–§—É–ª–ª –•–∞—É—Å'
        else:
            return '–°–µ—Ç'
    
    if 2 in c_cards:        
        if c_cards.count(2) == 2:
            return '–î–≤–µ –ø–∞—Ä—ã'
        else:
            return '–ü–∞—Ä–∞'
    
    return '–°—Ç–∞—Ä—à–∞—è –∫–∞—Ä—Ç–∞' 

solution('10 5 10 5 1')
```

## 7. –°–ª–∞–≤–Ω—ã–π –≥–æ—Ä–æ–¥
–ë–∞–≥–¥–∞–¥—Å–∫–∏–π –≤–æ—Ä –ø–µ—Ä–µ–º–µ—â–∞–µ—Ç—Å—è –ø–æ –≥–æ—Ä–æ–¥—É —Ä–∞–∑–º–µ—Ä–∞ N√óN, –ø–µ—Ä–µ–±–µ–≥–∞—è –ø–æ —Å–æ—Å–µ–¥–Ω–∏–º —Å–ø—Ä–∞–≤–∞ –∏–ª–∏ —Å–Ω–∏–∑—É –∫–ª–µ—Ç–∫–∞–º. –î–æ–±—Ä–∞—Ç—å—Å—è –¥–æ –∫–∞–∂–¥–æ–π –∏–∑ –∫–ª–µ—Ç–æ–∫ –≤–æ—Ä –º–æ–∂–µ—Ç –∑–∞ –∫–∞–∫–æ–µ-—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–µ–∫—É–Ω–¥ (–æ—Ç 1 –¥–æ 9), –∞ —Ç–∞–∫–∂–µ –Ω–∞ –∫–ª–µ—Ç–∫–µ –º–æ–∂–µ—Ç —Å—Ç–æ—è—Ç—å —Å—Ç—Ä–∞–∂–Ω–∏–∫, –≤ —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –∫–ª–µ—Ç–∫–∞ –±—É–¥–µ—Ç –æ–±–æ–∑–Ω–∞—á–∞—Ç—å—Å—è —á–∏—Å–ª–æ–º 0, –∏ –≤–æ—Ä –Ω–µ —Å–º–æ–∂–µ—Ç –ø—Ä–æ–π—Ç–∏ —á–µ—Ä–µ–∑ –Ω–µ–µ. –ï–º—É –Ω—É–∂–Ω–æ –¥–æ–±—Ä–∞—Ç—å—Å—è –∏–∑ —Å–∞–º–æ–π –ª–µ–≤–æ–π –≤–µ—Ä—Ö–Ω–µ–π —Ç–æ—á–∫–∏ (1;1) –≤ —Å–∞–º—É—é –Ω–∏–∂–Ω—é—é –ø—Ä–∞–≤—É—é —Ç–æ—á–∫—É (N;N).

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Ä–∏—Å—É–µ—Ç –º–∞—Ä—à—Ä—É—Ç, –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É –≤–æ—Ä —Å—É–º–µ–µ—Ç –ø—Ä–æ–π—Ç–∏ –≤–µ—Å—å –ø—É—Ç—å –∑–∞ –∫—Ä–∞—Ç—á–∞–π—à–µ–µ –≤—Ä–µ–º—è. –ï—Å–ª–∏ —Ç–∞–∫–æ–≥–æ –ø—É—Ç–∏ –Ω–µ—Ç, —Ç—Ä–µ–±—É–µ—Ç—Å—è –≤—ã–≤–µ—Å—Ç–∏ —Å–ª–æ–≤–æ Impossible.

```python
import numpy as np
       
class FameTown():

    def __init__(self, *rows):
        self.n = len(rows)
        self.max = 999
        self.distance = np.array([[int(r) for r in rows[i]] for i in range(self.n)])
        self.weight = np.full((self.n,self.n), self.max, dtype=int)
        self.field = np.full((self.n,self.n), '-', dtype='U')
        self.weight[0,0] = 0
        self.field[0,0] = '#'
    

    def solution(self):
        if self.distance[0,0] == 0 or self.distance[self.n - 1,self.n - 1] == 0:
            return 'Impossible'
        
        for i in np.arange(self.n):
            for j in np.arange(self.n):
                if i+1<self.n and self.distance[i+1,j] > 0:
                    if self.weight[i,j] + self.distance[i+1,j] < self.weight[i+1,j]:
                        self.weight[i+1,j] = self.weight[i,j] + self.distance[i+1,j]
                        
                if j+1<self.n and self.distance[i,j+1] > 0:
                    if self.weight[i,j] + self.distance[i,j+1] < self.weight[i,j+1]:
                        self.weight[i,j+1] = self.weight[i,j] + self.distance[i,j+1]
        
                
        i=j=self.n-1
        while self.weight[i,j] > 0:
            if self.weight[i,j] == self.max:
                return "Imposible"
            
            self.field[i,j] = '#'
            u=l=self.max
            if i-1>=0:
                u = self.weight[i-1,j]

            if j-1>=0:
                l = self.weight[i,j-1]

            if  l < u:
                j-=1
            else:
                i-=1
                    
        return '\n'.join([''.join(r) for r in self.field])

            
def test_input(inp):
    return  inp.splitlines()

# print(FameTown(*['1112', '1001', '1101', '0111']).solution())


```

    deleted, –µ—Å–ª–∏ –∫–ª—é—á –±—ã–ª –≤ –ø–µ—Ä–≤–æ–º —Å–ª–æ–≤–∞—Ä–µ, –Ω–æ –µ–≥–æ –Ω–µ –±—ã–ª–æ –≤–æ –≤—Ç–æ—Ä–æ–º;
    changed, –µ—Å–ª–∏ –∫–ª—é—á –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –≤ –æ–±–æ–∏—Ö —Å–ª–æ–≤–∞—Ä—è—Ö, –Ω–æ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è;
    equal, –µ—Å–ª–∏ –∫–ª—é—á –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –≤ –æ–±–æ–∏—Ö —Å–ª–æ–≤–∞—Ä—è—Ö, –∏ –∑–Ω–∞—á–µ–Ω–∏—è —Å–æ–≤–ø–∞–¥–∞—é—Ç.


```python
def key_difference(dict1, dict2):
    res = {}
    for k in dict1:
        if k in dict2:
            res[k] = 'equal' if dict1[k] == dict2[k] else 'changed'
        else:
            res[k] = 'deleted'
            
    for k in dict2:
        if k not in dict1:
            res[k] = 'added'
    return res

key_difference({
    'one': 'eon',
    'two': 'two',
    'four': 'True'
}, {
    'two': 'own',
    'zero': '4',
    'four': 'True'
})
```

## 3. –ë–∞–±–∫–∞–≥–µ–¥–¥–æ–Ω
–•–æ–≤–∞—Ä–¥ –∏ –í–∏–Ω—Å —É–±–µ–≥–∞—é—Ç –æ—Ç –¥–µ–º–æ–Ω–∞ –∑–ª–æ–π –±–∞–±—É—à–∫–∏, –∫–æ—Ç–æ—Ä–æ–≥–æ –æ–Ω–∏ —Å–ª—É—á–∞–π–Ω–æ –ø—Ä–∏–∑–≤–∞–ª–∏. –†–∏—Ç—É–∞–ª –ø—Ä–æ–∏–∑–æ—à–µ–ª –Ω–∞ –Ω—É–ª–µ–≤—ã—Ö –ª–∏–Ω–µ–π–Ω—ã—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö, –∏ –ø–µ—Ä–µ–¥ —Ç–µ–º –∫–∞–∫ –±–∞–±—É—à–∫–∞ –ø–æ–≥–Ω–∞–ª–∞—Å—å –∑–∞ –Ω–∏–º–∏, –•–æ–≤–∞—Ä–¥ –æ—Ç–±–µ–∂–∞–ª –Ω–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É n, –∞ –í–∏–Ω—Å –Ω–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É m, –ø–æ—Å–ª–µ —á–µ–≥–æ –æ–Ω–∏ –ø—Ä–æ–¥–æ–ª–∂–∏–ª–∏ —É–±–µ–≥–∞—Ç—å –æ—Ç –Ω—É–ª–µ–≤–æ–π –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Å –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π —Å–∫–æ—Ä–æ—Å—Ç—å—é, –∞ –±–∞–±—É—à–∫–∞ –Ω–∞—á–∞–ª–∞ —Å–≤–æ–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤ –¥–≤–∞ —Ä–∞–∑–∞ –±—ã—Å—Ç—Ä–µ–µ –Ω–∏—Ö.
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –Ω—É–∂–Ω–æ –ø—Ä–æ–±–µ–∂–∞—Ç—å –±–∞–±—É—à–∫–µ, —á—Ç–æ–±—ã –Ω–∞—Å—Ç–∏–≥–Ω—É—Ç—å –•–æ–≤–∞—Ä–¥–∞ –∏ –í–∏–Ω—Å–∞.

```python
def solution(inputmn):
    sign = lambda x: (1, -1)[x < 0]
    x0 = 0
    x = [int(_) for _ in inputmn.split(' ')]
    if abs(x0 - x[0]) <= abs(x0 - x[1]):
        n, m = x
    else:
        m, n = x

    x01 = 2 * n - x0
    x21 = m + sign(m) * abs(n - x0)
    x02 = x01 + 2 * abs(x21 - x01) * sign(m)
    return int(abs(x02 - x01) + abs(x01))


solution('-1 1')
```

## 4. –°–∞–º–æ–µ —Å–∫—É—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ

## 5. –î—Ä—É–∂–µ–ª—é–±–Ω—ã–π –≥–æ—Ä–æ–¥–æ–∫
–í –≥–æ—Ä–æ–¥–∫–µ BeeGeek, —á—Ç–æ–±—ã –≤—Å–µ –∂–∏—Ç–µ–ª–∏ –≤–∏–¥–µ–ª–∏—Å—å –∫–∞–∫ –º–æ–∂–Ω–æ —á–∞—â–µ, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –≤—Å–µ–≥–æ –æ–¥–Ω–∞ —É–ª–∏—Ü–∞, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∞—è —Å–æ–±–æ–π –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—É—é –ø—Ä—è–º—É—é. –ù–∞ –Ω–µ–π –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –∫–∞–∫–æ–µ-—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–æ–º–æ–≤, —É –∫–∞–∂–¥–æ–≥–æ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö, –µ—Å—Ç—å —Å–≤–æ—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω–∞—è –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º, –∏ —Å–≤–æ–π —Ü–≤–µ—Ç (—Ü–≤–µ—Ç–∞ –¥–æ–º–æ–≤ –º–æ–≥—É—Ç –ø–æ–≤—Ç–æ—Ä—è—Ç—å—Å—è).
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–æ–º–∞–º–∏ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–≥–æ —Ü–≤–µ—Ç–∞. –ï—Å–ª–∏ –¥–æ–º–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–≥–æ —Ü–≤–µ—Ç–∞ –Ω–µ—Ç, —Ç–æ —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ 0.

```python
import pandas as pd
def solution(coords, colors):
    ser = pd.Series(coords.split(), index = colors.split()).astype(int)
    return ser.groupby(level=0).apply(lambda x: x.max() - x.min() if len(x) % 2 == 0 else 0).max()

solution('2 5 10 11 16 17 18 20', 'r w g w g r w b')
        
```

```python
def solution(coords, colors):
    coords = list(map(int, coords.split()))
    colors = colors.split()
    res = 0
    for i in range(0, len(colors) // 2 + 1):
        for j in range(len(colors) - 1, len(colors) // 2, -1):
            if colors[i] == colors[j]:
                res = max(res, coords[j] - coords[i])
                break
    return res


solution('1 3 5 6 10 21 22', 'r g b w w r g')
```

## 6. –ü–æ–≤—Ç–æ—Ä—è–π–ø–æ–≤—Ç–æ—Ä—è–π–ø–æ–≤—Ç–æ—Ä—è–π
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –Ω–∞ –∫–∞–∫–æ–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö –ø–æ–¥—Å—Ç—Ä–æ–∫ –º–æ–∂–Ω–æ —Ä–∞–∑–±–∏—Ç—å –¥–∞–Ω–Ω—É—é —Å—Ç—Ä–æ–∫—É —Ç–∞–∫, —á—Ç–æ–±—ã –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å—Ç—Ä–æ–∫–∏ –±—ã–ª–∏ –∑–∞–¥–µ–π—Å—Ç–≤–æ–≤–∞–Ω—ã.

```python
# %%timeit
from collections import Counter
import numpy as np

def solution(ss):
    x = np.array(list(Counter(ss).values()))
    for i in range(x.min(), 0, -1):
        if np.all(x % i == 0):
            return i
    

solution('–ø–æ–≤—Ç–æ—Ä—è–π–ø–æ–≤—Ç–æ—Ä—è–π–ø–æ–≤—Ç–æ—Ä—è–π')
```

```python
def solution(ss):
    subss = ''
    for s in ss:
        subss += s
        sc = ss.count(subss)
        if sc == 1 or len(ss) / len(subss) == sc:
            return sc
        
solution('bbabbbab')
```

## 7. –£–º–Ω–æ–∂–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫
–ù–∞–∑–æ–≤–µ–º —É–º–Ω–æ–∂–µ–Ω–∏–µ–º —Å—Ç—Ä–æ–∫–∏ –Ω–∞ —á–∏—Å–ª–æ –∑–∞–ø–∏—Å—å –≤ —Ñ–æ—Ä–º–∞—Ç–µ n(string), –≥–¥–µ n ‚Äî –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –∞ string ‚Äî —Å—Ç—Ä–æ–∫–∞, –∫–æ—Ç–æ—Ä–∞—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –∑–∞–ø–∏—Å–∞–Ω–∞ n —Ä–∞–∑. –†–∞—Å–∫—Ä—ã—Ç–∏–µ–º —É–º–Ω–æ–∂–µ–Ω–∏—è –±—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—ã–π –≤–∞—Ä–∏–∞–Ω—Ç –¥–∞–Ω–Ω–æ–π –∑–∞–ø–∏—Å–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä, —Å—Ç—Ä–æ–∫–∞ ti2(Be)3(Ge) –ø–æ—Å–ª–µ —Ä–∞—Å–∫—Ä—ã—Ç–∏—è –≤ –Ω–µ–π –≤—Å–µ—Ö —É–º–Ω–æ–∂–µ–Ω–∏–π –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥ tiBeBeGeGeGe.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞—Å–∫—Ä—ã–≤–∞–µ—Ç –≤—Å–µ —É–º–Ω–æ–∂–µ–Ω–∏—è –≤ —Ç–µ–∫—Å—Ç–µ –∏ –≤—ã–≤–æ–¥–∏—Ç –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

```python
import re


def solution(ss):
    _re_brace = re.compile(r'(\d+)\(([a-z]+)\)', re.ASCII)
    while _re_brace.search(ss):
        ss = _re_brace.sub(lambda x: int(x.group(1)) * x.group(2), ss)
    return ss


solution('0(s)he0(be)lie0(ve)d')
```

```python
# %%timeit
import numpy as np


def solution(ls):
    a = np.array(ls.split(' '), dtype=int)
    a=np.random.randint(100,size=20)
    print(a)
    d = np.ones(a.size, dtype=int)
    p = np.full(a.size, fill_value=-1, dtype=int)
    for i in range(d.size):
        for j in range(i):
            if a[j] < a[i]:
                d[i] = max(d[i], 1 + d[j])
                p[i] = j
    print(d)
    pos = np.argmax(d)
    print(pos)
    print(p[pos])
    d.fill(0)
    while pos != -1:
        d[pos] = 1
        pos = p[pos]

    
    return ' '.join(a[d.astype(bool)].astype(str))


solution('2 5 3 4 6 1')
```

```python
# %%timeit
import numpy as np


def solution(ls):
    a = np.array(ls.split(), dtype=int)
    fsum = a.sum()
    sum2, m = np.divmod(fsum, 2)
    print(f"{fsum=}")
    if m == 1:
        return False
    s2, polusum, c, rem = sum2, 0, 0, fsum - sum2
    while s2 >= a.min():
        to_check = np.array([s2,rem])
#         polusum += np.sum(to_check[np.isin(to_check, a)]) * 2**c
        print(f"{s2=},{m=},{polusum=},{rem=},{c=}")
        print(to_check[np.isin(to_check, a)])

#         to_check = [s2, s2+m, s2-m]
        if np.sum(np.isin(to_check,a)) > 1:
            return True
        
        s2, m = np.divmod(rem, 2)
        rem -= s2
        
        c += 1
    return False


print(1)
print(solution('4 6'))
print(2)
print(solution('1 2 3 4 6'))
print(3)
print(solution('4 2 5 2 7 3 9 3 6 2 3 43 1 44 123 1'))
print(4)
print(solution('2 3 4 5'))
print(5)
print(solution('2 7 5'))
```
